```python
import folium 
# Memasukkan data ke dalam DataFrame
df = pd.DataFrame(volbypost)

# Membuat peta Folium yang terpusat di Indonesia
map_center = [-2.5, 117.5]  # Titik tengah Indonesia
m = folium.Map(location=map_center, zoom_start=5)

# Menambahkan marker (penanda) untuk setiap kota
for i, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['kecamatan'],
        icon=folium.Icon(icon="cloud")
    ).add_to(m)

# Menampilkan peta
m.save("indonesia_map.html")
m
```
